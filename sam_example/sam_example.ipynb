```python
def core_list(process):
    n = len(process[0])
    core = []
    for row in range(n):
        for col in range(n):
            if process[row][col] == '1':
                if col != 0 and row != 0:
                    core.append([row, col])
    return core
```

```python
def check_line(process, core):
    core_check = [[0,0,0,0] for row in range(len(core))]
    n = len(process[0])
    for len_core in range(len(core)):
        row_core = core[len_core][0]
        col_core = core[len_core][1]
        
        #왼쪽
        for col in range(col_core):
            if process[row_core][col] == '0':
                core_check[len_core][0] +=1
            else:
                core_check[len_core][0] = 0
                
                break
        #오른쪽
        for col in range(col_core+1,n):
            if process[row_core][col] == '0':
                core_check[len_core][1] +=1
            else:
                core_check[len_core][1] = 0
                
                break
        #위       
        for row in range(row_core):
            if process[row][col_core] == '0':
                core_check[len_core][2] +=1
            else:
                core_check[len_core][2] = 0
                break
        #아래
        for row in range(row_core+1,n):
            if process[row][col_core] == '0':
                core_check[len_core][3] +=1
            else:
                core_check[len_core][3] = 0
                break
            
                    
    return core_check
                
```

```python
def route_check(process, core, core_check):
    count = 0
    for row_core, col_core in core:
        min(core_check[count])
        
        count+=1
        
```
