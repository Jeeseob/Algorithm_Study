```python
M = 6
array = [19,15,10,17]
```

```python
def cutRicecake(array,M):
    ricecakes = sorted(array)
    target = ricecakes[-1]//2
    while True:
        print(target)
        midIndex = binary_search(0,len(ricecakes)-1, ricecakes, target)
        #print(midIndex)
        answer = find_cut(array, midIndex)
        
        if answer == M:
            return midIndex
        elif answer < M:
            target = target//2
        else:
            target += target//2
            
    return midIndex
```

```python
def binary_search(start, end, array, target):
    if start == end:
        return start
    
    mid = (start+end)//2
    
    if array[mid] == target:
        return mid
    elif array[mid] > target:
        return binary_search(start, mid-1, array, target)
    else:
        return binary_search(mid+1, end, array, target)
```

```python
def find_cut(array,midIndex):
    result = 0
    find_array = array[midIndex:]
    for cut in find_array:
         result += (cut - array[midIndex]) 
    return result
    
```
